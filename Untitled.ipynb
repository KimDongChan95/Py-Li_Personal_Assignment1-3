```python
def calculate_arrest_rate(arrested, occurred):
    if occurred == 0:
        return 0  # 발생 건수가 0이면 검거율도 0
    return round((arrested / occurred) * 100, 2)
```

```python
df['살인 검거율 (%)'] = df.apply(lambda row: calculate_arrest_rate(row['살인(검거)'], row['살인(발생)']), axis=1)
df['강도 검거율 (%)'] = df.apply(lambda row: calculate_arrest_rate(row['강도(검거)'], row['강도(발생)']), axis=1)
df['강간 검거율 (%)'] = df.apply(lambda row: calculate_arrest_rate(row['강간(검거)'], row['강간(발생)']), axis=1)
df['절도 검거율 (%)'] = df.apply(lambda row: calculate_arrest_rate(row['절도(검거)'], row['절도(발생)']), axis=1)
df['폭력 검거율 (%)'] = df.apply(lambda row: calculate_arrest_rate(row['폭력(검거)'], row['폭력(발생)']), axis=1)
```
